## VI. Data Cleaning & Feature Selection
### Usually these two process don't go together... but since we are seeing some columns containing too many NAs, our data cleaning might actually be cleaning out colunns instead of rows.

```python
NA_count_table = pd.DataFrame(columns = ['Feature', 'NA_count'])
NA_count_table['Feature'] = data.columns
for i in range(len(data.columns)):
    NA_count_table.loc[i, 'NA_count'] = sum(pd.isna(data[data.columns[i]]))
NA_count_table.sort_values("NA_count", ascending = False)
```

### We drop the columns with NA_count over 100.

```python
Features_to_remove = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'LotFrontage']
data_clean = data.drop(Features_to_remove, axis = 1)
```

### We can also remove rows that have NA. This might not be the best way to clean it, but let's start with an easy approach:

```python
data_clean.dropna(axis = 0, how = "any", inplace = True)
```

### How much data did we lose?

```python
print(data.shape)
print(data_clean.shape)
#Notbad
```

```python
#double check
np.array(pd.isna(data_clean).values).flatten().sum()
```

## VII. Data Preparation
### The dataset is still not ready to be modeled yet. Categorical data can't be used in a regression problem.  We can try label encode the categorical columns.


```python
from sklearn.preprocessing import LabelEncoder
def dummyEncode(df):
        columnsToEncode = list(df.select_dtypes(include=['category','object']))
        le = LabelEncoder()
        for feature in columnsToEncode:
            try:
                df[feature] = le.fit_transform(df[feature])
            except:
                print('Error encoding '+feature)
        return df
```

```python
data_clean_encoded = dummyEncode(data_clean)
```

### Last step! Split X (Features) and y (SalePrice) to train and test set.

```python
from sklearn.model_selection import train_test_split
X = data_clean_encoded.drop(["SalePrice"], axis = 1)
y = data_clean_encoded["SalePrice"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
data_clean_encoded
```

## VIII. Modeling
### Now we have all we need to model. We will start with a easy model - Linear Regression.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

### Let's visualize how bad our model does.

```python
plt.scatter(y_test, y_pred)
plt.xlim(0, 800000);
plt.ylim(0, 800000);
plt.xlabel("Actual");
plt.ylabel("Predicted");
plt.title("NotBad")
plt.plot([0,1000000], [0,1000000], linestyle = "--", color = "red", alpha = 0.5)
```

```python
#rmse = mean_squared_error(y_test, y_pred)**0.5
import math
y_test_log = [math.log(item) for item in y_test]
y_pred_log = [math.log(item) for item in y_pred]
mean_squared_error(y_test_log, y_pred_log)**0.5
```
